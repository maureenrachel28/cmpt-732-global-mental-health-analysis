```python
import pandas as pd  
import requests
from bs4 import BeautifulSoup
```

```python
def lastWord(string):
   
    # split by space and converting 
    # string to list and
    lis = list(string.split(" "))
#     lis = (string.split(" "))

     
    # length of list
    
    length = len(lis)
     
    # returning last element in list
    return " ".join(lis[3:length-1]) 
 
 
# Driver code
string = "Mental Health Atlas Democratic Republic of the Congo 2020"
print(lastWord(string))
```

```python
# making dataframe  
# df = pd.read_excel(links)  
df = pd.read_csv(links)  


# output the dataframe 
df
```

```python
df['country'] = df['country'].apply(lambda x: lastWord(x))
df
```

```python
for ind in df.index:
    url = df['country-href'][ind]
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all hyperlinks present on webpage
    links = soup.find_all('a')

    i = 0
    # From all links check for pdf link and
    # if present download file
    for link in links:
        if ('.pdf' in link.get('href', [])):
            i += 1
#             print("Downloading file: ", i)

            # Get response object for link
            response = requests.get(link.get('href'))

            # Write content in pdf file
            pdf = open(df['country'][ind]+".pdf", 'wb')
            pdf.write(response.content)
            pdf.close()
            print(df['country'][ind]," downloaded")

#     print("All PDF files downloaded")
```
